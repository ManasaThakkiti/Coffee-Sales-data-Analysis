```python
#Convert the hour_of_day column to integer type (handle strings like "twelve").¶
word_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5,"six": 6, "seven": 7, "eight": 8, "nine": 9, "ten": 10, 
               "eleven": 11, "twelve": 12}
df["hour_of_day"]=df["hour_of_day"].replace(word_to_num)
df["hour_of_day"]=pd.to_numeric(df["hour_of_day"],errors="coerce")
df["hour_of_day"]=df["hour_of_day"].astype("Int64")
print(df["hour_of_day"].head())
```

```python
# convert the money columns to numerics
df["money"]=pd.to_numeric(df["money"],errors="coerce")
```

```python
# --- Numeric columns ---
numeric_cols = ["hour_of_day", "money", "Weekdaysort", "Monthsort"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df[col].astype(float)  # convert to float before filling
    df[col] = df[col].fillna(df[col].mean())

# --- Categorical columns ---
categorical_cols = ["cash_type", "coffee_name", "Time_of_Day", "Weekday", "Month_name"]
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

# --- Date column ---
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].fillna(pd.Timestamp('1970-01-01'))

# --- Time column ---
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
df['Time'] = df['Time'].fillna(datetime.time(12, 0, 0))
```

```python
def remove_outliers_IQR(group):
    Q1 = group['money'].quantile(0.25)
    Q3 = group['money'].quantile(0.75)
    IQR = Q3 - Q1
    return group[(group['money'] >= Q1 - 1.5 * IQR) & (group['money'] <= Q3 + 1.5 * IQR)]

# Apply this to each Time_of_Day group
df_no_outliers = df.groupby('Time_of_Day', group_keys=False).apply(remove_outliers_IQR)
# Replot
plt.figure(figsize=(4,4))
sns.boxplot(data=df_no_outliers, x='Time_of_Day', y='money')
plt.title('Spending Distribution After Removing Outliers', fontsize=8)
plt.xlabel('Time of Day')
plt.ylabel('Money Spent ($)')
plt.tight_layout()
plt.show()
```
